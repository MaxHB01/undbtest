```python
orderItemAenc = pd.read_sql_query("select * from sales_order_item", engine_ae)
orderAenc = pd.read_sql_query("select * from sales_order", engine_ae)

orderItemNorthwind = pd.read_sql_query("select * from [Order Details]", engine_nw)
orderNorthwind = pd.read_sql_query("select * from Orders", engine_nw)

orderItemAdventrue = pd.read_sql_query(f"select * from [AdventureWorks2019].[Sales].[SalesOrderDetail]", engine_aw)
orderAdventure = pd.read_sql_query(f"select * from [AdventureWorks2019].[Sales].[SalesOrderHeader]", engine_aw)
territoryAdventure = pd.read_sql_query(f"select * from [AdventureWorks2019].[Sales].[SalesTerritory]", engine_aw)

# sql_query = "SELECT name FROM sqlite_master WHERE type='table' AND name LIKE 'Sales%';"
# print(pd.read_sql_query(sql_query, adventure_connection))

engine_nw.dispose()
```

```python
# Assuming df1 is sales_order_item and df2 is sales_order
df1 = orderItemAenc  # your sales_order_item dataframe
df2 = orderAenc  # your sales_order dataframe

# Merge the dataframes on 'id'
merged_df = pd.merge(df1, df2, on='id')

# Select the desired columns
selected_df = merged_df[['id', 'prod_id', 'cust_id', 'sales_rep', 'order_date', 'region', 'quantity', 'ship_date']]

# Rename the columns
order_df = selected_df.rename(columns={
    'id': 'Id',
    'prod_id': 'ProductId',
    'cust_id': 'CusromerId',
    'sales_rep': 'EmployeeId',
    'order_date': 'OrderDate',
    'region': 'Region',
    'quantity': 'Quantity',
    'ship_date': 'ShipDate'
})

order_df.insert(1, 'Database', 'AE')

order_aenc_df = order_df
order_aenc_df
```

```python
# Assuming df1 is sales_order_item and df2 is sales_order
df1 = orderItemNorthwind  # your sales_order_item dataframe
df2 = orderNorthwind  # your sales_order dataframe


# Merge the dataframes on 'id'
merged_df = pd.merge(df1, df2, on='OrderID')

# Select the desired columns
selected_df = merged_df[['OrderID', 'ProductID', 'CustomerID', 'EmployeeID', 'OrderDate', 'ShipRegion', 'Quantity', 'ShippedDate']]

# Rename the columns
order_df = selected_df.rename(columns={
    'OrderID': 'Id',
    'ProductID': 'ProductId',
    'CustomerID': 'CusromerId',
    'EmployeeID': 'EmployeeId',
    'OrderDate': 'OrderDate',
    'ShipRegion': 'Region',
    'Quantity': 'Quantity',
    'ShippedDate': 'ShipDate'
})

order_df.insert(1, 'Database', 'NW')

order_northwind_df = order_df
order_northwind_df
```

```python
# Assuming df1 is sales_order_item and df2 is sales_order
df1 = orderItemAdventrue  # your sales_order_item dataframe
df2 = orderAdventure  # your sales_order dataframe
df3 = territoryAdventure

# Merge the dataframes on 'id'
merged_df = pd.merge(df1, df2, on='SalesOrderID')
merged_df = pd.merge(merged_df, df3, on='TerritoryID')

# Select the desired columns
selected_df = merged_df[['SalesOrderID', 'ProductID', 'CustomerID', 'SalesPersonID', 'OrderDate', 'Name', 'OrderQty', 'ShipDate']]

# Rename the columns
order_df = selected_df.rename(columns={
    'SalesOrderID': 'Id',
    'ProductID': 'ProductId',
    'CustomerID': 'CusromerId',
    'SalesPersonID': 'EmployeeId',
    'OrderDate': 'OrderDate',
    'Name': 'Region',
    'OrderQty': 'Quantity',
    'ShipDate': 'ShipDate'
})

order_df.insert(1, 'Database', 'AW')

order_adventrue_df = order_df
order_adventrue_df
```
